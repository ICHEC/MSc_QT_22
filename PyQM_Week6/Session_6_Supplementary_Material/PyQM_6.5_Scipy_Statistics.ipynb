```python
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as rnd

import scipy
from scipy import stats
```

## Distributions

There are many distributions available with `scipy.stats`. A full List can be found at [here](https://docs.scipy.org/doc/scipy/reference/reference/stats.html#module-scipy.stats). The following can be extracted once the distribution is created:
- `.pdf`: **P**robability **D**istribution **F**unction,
- `.pmf`: **P**robability **M**ass **F**unction,
- `.cdf`: **C**umlative **D**istribution **F**unction,
- `.rvs`: **R**andom **V**alue**s**,
- `.mean()`: **Mean**,
- `.std()`: **St**andard **D**eviation,
- `.var()`: **Var**iance.

### Normal Distribution

A normal distribution with a mean ($\mu$), `loc` and standard deviation ($\sigma$), `scale` is created with `scipy.stats.norm(loc, scale)`, according to the function
$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2} \left( \frac{x-\mu}{\sigma} \right)^2}.$$

```python
D = stats.norm(loc = 10, scale=.5)

x = np.linspace(D.mean()-3*D.std(),D.mean()+3*D.std(),100)

fig = plt.figure()
ax = fig.add_axes([.1,.1,.8,.8])

ax.hist(D.rvs(size=1000), bins=50, density=True, label = 'data')
ax.plot(x, D.pdf(x), ls = ':', label = 'pdf')
ax.plot(x, D.cdf(x), ls = '--', label = 'cdf')

ax.set_xlim(x[0],x[-1])

ax.set_title('Normal Distribution\n$\mu='+str(D.mean())+'$, $\sigma='+str(D.std())+'$, and $\sigma^2='+str(D.var())+'$')
ax.legend()
plt.show()
```

### Poisson Distribution

A Poisson distribution with a mean ($\lambda$), `loc` is created with `scipy.stats.poisson(lambda)`, according to the function
$$f(x) = \frac{\lambda^x}{x!} e^{-\lambda}$$

```python
D = stats.poisson(3)

n = np.arange(0,15)
bin_array = np.arange(0,10,1)

fig = plt.figure()
ax = fig.add_axes([.1,.1,.8,.8])

ax.hist(D.rvs(size=100), bins=bin_array, density=True, label = 'data')
ax.step(n, D.pmf(n), ls = ':', label = 'pmf')
ax.step(n, D.cdf(n), ls = '--', label = 'cdf')

ax.set_title('Poisson Distribution\n$\mu='+str(D.mean())+'$, $\sigma='+str(np.round(D.std(),3))+'$, and $\sigma^2='+str(D.var())+'$')
ax.legend()
plt.show()
```
